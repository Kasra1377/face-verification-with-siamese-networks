```python
import cv2
import numpy as np
from imutils import paths
from sklearn.utils import shuffle
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
# define a path for data and list all of the images
# in the path and then shuffle them
path = "extracted-celeb-data"
imagePaths = list(paths.list_images(path))
imagePaths = shuffle(imagePaths)
```

```python
# initialize the data list
data = []

# loop through the list of image paths
for i in imagePaths:
  image = cv2.imread(i , 0)
  image = img_to_array(image)
  data.append(image)
```

```python
# convert the list into Numpy array format
data = np.array(data , dtype = "float32")
```

```python
np.random.seed(42)
# create a set of random numbers between 0 and the
# size of the dataset with a size of 25
samples = np.random.choice(np.arange(0 , len(data)), size=(25,))

# loop through random numbers, and save the selected images
for i in samples:
  cv2.imwrite("samples/image-{}.png".format(i), data[i]) 
```
